```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import csv
import re
```

```python
from selenium import webdriver

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

driver = webdriver.Chrome(options=chrome_options)
```

## 크롤링

```python
driver.get("https://likms.assembly.go.kr/bill/main.do")
```

```python
# '검색' 버튼 클릭
search_button = driver.find_element(By.XPATH, '//button[@class="btnSch01"]')
search_button.click()

# CSV 파일을 위한 설정
csv_file = open("scraped_data.csv", "w", newline='', encoding='utf-8')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Column 1", "Column 2", "..."])  # 이 부분은 실제 테이블의 열 이름으로 바꾸어 주세요

page_num = 1
while True:
    # 페이지가 로드되기를 기다림
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, '//div[@class="tableCol01"]/table/tbody'))
    )

    # 테이블 셀의 데이터를 배열에 저장
    rows = driver.find_elements(By.XPATH, '//div[@class="tableCol01"]/table/tbody/tr')
    for i, row in enumerate(rows):
        cells = row.find_elements(By.TAG_NAME, "td")
        cell_data = [cell.text for cell in cells]

        img_elem = row.find_element(By.TAG_NAME, "img")
        alt_value = img_elem.get_attribute("alt")
        cell_data.append(alt_value)

        href_elem = row.find_element(By.TAG_NAME,"a")
        href_value = href_elem.get_attribute("href")
        pattern = r"javascript:fGoDetail\('(.*?)', 'billSimpleSearch'\)"
        matches = re.search(pattern, href_value)

        if matches:
            extracted = matches.group(1)
            cell_data.append(extracted)
        csv_writer.writerow(cell_data)

    print(f"페이지 {page_num} 저장 완료.")

    # 다음 페이지로 이동
    page_num += 1
    driver.execute_script(f"GoPage({page_num})")

    # 약간의 시간을 주어 페이지 로딩을 기다림
    time.sleep(2)

    # 특정 조건에 도달하면 루프 종료 (예: 마지막 페이지)
    if page_num > 2538:  # 이 부분은 실제 마지막 페이지 번호로 바꾸어 주세요
        print("마지막 페이지에 도달했습니다.")
        break


```

```python
# 웹 드라이버 종료
driver.quit()

# CSV 파일 닫기
csv_file.close()
```
